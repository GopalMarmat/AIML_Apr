```python
#float variable
a=25.12 #Float variable

def calculator(a, b, condition):
    print("a variable types",type(a))
    print("b variable types",type(b))
    if condition == "add": #this conditon for addition
        return a+b
    elif condition=="sub": #this conditon for substraction
        return a-b
    elif condition == "mul": #this conditon for mutliplication
        return a*b
    elif condition == "div": #this conditon for division
        return a/b
    else:
        print("Please pass proper condition")
result=calculator(25.8,12,"add")
print(result)
    
```

## String Inbuilt Fuction

| **Method**              | **Description**                        | **Application**                                                       | **Example**                                                                          |
| ----------------------- | -------------------------------------- | --------------------------------------------------------------------- | ------------------------------------------------------------------------------------ |
| `capitalize()`          | Capitalizes first character            | To capitalize the first letter of a string                            | **Input:** `"hello".capitalize()` <br> **Output:** `Hello`                           |
| `casefold()`            | Aggressive lowercase (for comparisons) | Useful for case-insensitive string comparisons                        | **Input:** `"ß".casefold()` <br> **Output:** `ss`                                    |
| `center(width, fill)`   | Center aligns string                   | For centering a string within a specified width                       | **Input:** `"hi".center(10, '-')` <br> **Output:** `--hi-------`                     |
| `count(sub)`            | Count occurrences of substring         | To count how many times a substring appears in a string               | **Input:** `"banana".count("a")` <br> **Output:** `3`                                |
| `encode()`              | Encode string to bytes                 | Convert a string to its byte representation                           | **Input:** `"hi".encode()` <br> **Output:** `b'hi'`                                  |
| `endswith(suffix)`      | Checks string end                      | To check if a string ends with a specific suffix                      | **Input:** `"hello".endswith("lo")` <br> **Output:** `True`                          |
| `expandtabs(tabsize)`   | Replace `\t` with spaces               | For converting tab characters to spaces in a string                   | **Input:** `"a\tb".expandtabs(4)` <br> **Output:** `a   b`                           |
| `find(sub)`             | Return index of substring, or -1       | To find the first occurrence of a substring in a string               | **Input:** `"hello".find("l")` <br> **Output:** `2`                                  |
| `format()`              | Format string                          | To format a string with placeholders and variables                    | **Input:** `"{} {}".format("Hi","All")` <br> **Output:** `Hi All`                    |
| `format_map(mapping)`   | Format using a dict                    | Format a string using a dictionary mapping                            | **Input:** `"{name}".format_map({"name":"Bob"})` <br> **Output:** `Bob`              |
| `index(sub)`            | Like `find()` but errors if not found  | To find the index of a substring, raises ValueError if not found      | **Input:** `"hello".index("e")` <br> **Output:** `1`                                 |
| `isalnum()`             | Alphanumeric?                          | To check if a string contains only alphanumeric characters            | **Input:** `"abc123".isalnum()` <br> **Output:** `True`                              |
| `isalpha()`             | Alphabetic?                            | To check if a string contains only alphabetic characters              | **Input:** `"abc".isalpha()` <br> **Output:** `True`                                 |
| `isascii()`             | ASCII characters only?                 | To check if all characters in the string are ASCII characters         | **Input:** `"π".isascii()` <br> **Output:** `False`                                  |
| `isdecimal()`           | Decimal digits?                        | To check if the string consists of decimal digits only                | **Input:** `"123".isdecimal()` <br> **Output:** `True`                               |
| `isdigit()`             | Digits? (includes superscripts etc.)   | To check if the string consists of digits (including special symbols) | **Input:** `"²".isdigit()` <br> **Output:** `True`                                   |
| `isidentifier()`        | Valid Python identifier?               | To check if a string is a valid Python variable name                  | **Input:** `"var1".isidentifier()` <br> **Output:** `True`                           |
| `islower()`             | All letters lowercase?                 | To check if all letters in a string are lowercase                     | **Input:** `"hello".islower()` <br> **Output:** `True`                               |
| `isnumeric()`           | Numeric?                               | To check if a string contains numeric characters, including fractions | **Input:** `"Ⅷ".isnumeric()` <br> **Output:** `True`                                 |
| `isprintable()`         | Printable characters?                  | To check if all characters in a string are printable                  | **Input:** `"hello".isprintable()` <br> **Output:** `True`                           |
| `isspace()`             | Only whitespace?                       | To check if the string contains only spaces or tab characters         | **Input:** `"   ".isspace()` <br> **Output:** `True`                                 |
| `istitle()`             | Title case?                            | To check if the string is in title case                               | **Input:** `"Hello World".istitle()` <br> **Output:** `True`                         |
| `isupper()`             | All uppercase?                         | To check if all letters in a string are uppercase                     | **Input:** `"HELLO".isupper()` <br> **Output:** `True`                               |
| `join(iterable)`        | Join iterable into string              | To join elements of an iterable with a separator string               | **Input:** `" ".join(["a","b"])` <br> **Output:** `a b`                              |
| `ljust(width, fill)`    | Left align                             | To left-align a string within a specified width                       | **Input:** `"hi".ljust(10, '-')` <br> **Output:** `hi--------`                       |
| `lower()`               | Lowercase                              | To convert all characters in a string to lowercase                    | **Input:** `"HELLO".lower()` <br> **Output:** `hello`                                |
| `lstrip(chars)`         | Strip left side                        | To remove leading whitespace or specified characters                  | **Input:** `"  hi".lstrip()` <br> **Output:** `hi`                                   |
| `maketrans()`           | Create translation table               | To create a character translation map                                 | **Input:** `str.maketrans("abc", "123")` <br> **Output:** `{97: 49, 98: 50, 99: 51}` |
| `partition(sep)`        | Split into 3 parts                     | To split a string into three parts around a separator                 | **Input:** `"a-b".partition("-")` <br> **Output:** `('a', '-', 'b')`                 |
| `removeprefix(prefix)`  | Remove prefix if present               | To remove a prefix from a string, if it exists                        | **Input:** `"HelloWorld".removeprefix("Hello")` <br> **Output:** `World`             |
| `removesuffix(suffix)`  | Remove suffix if present               | To remove a suffix from a string, if it exists                        | **Input:** `"file.txt".removesuffix(".txt")` <br> **Output:** `file`                 |
| `replace(old, new)`     | Replace substring                      | To replace all occurrences of a substring                             | **Input:** `"hi hi".replace("hi", "bye")` <br> **Output:** `bye bye`                 |
| `rfind(sub)`            | Rightmost index or -1                  | To find the rightmost occurrence of a substring                       | **Input:** `"hello".rfind("l")` <br> **Output:** `3`                                 |
| `rindex(sub)`           | Rightmost index, errors if not found   | To find the rightmost occurrence, raises error if not found           | **Input:** `"hello".rindex("l")` <br> **Output:** `3`                                |
| `rjust(width, fill)`    | Right align                            | To right-align a string within a specified width                      | **Input:** `"hi".rjust(10, '-')` <br> **Output:** `-------hi`                        |
| `rpartition(sep)`       | Rightmost partition                    | To split a string from the right, into 3 parts                        | **Input:** `"a-b-c".rpartition("-")` <br> **Output:** `('a', '-', 'b-c')`            |
| `rsplit(sep, maxsplit)` | Split from right                       | To split a string from the right, limiting splits                     | **Input:** `"a,b,c".rsplit(",", 1)` <br> **Output:** `['a,b', 'c']`                  |
| `rstrip(chars)`         | Strip right side                       | To remove trailing whitespace or specified characters                 | **Input:** `"hi   ".rstrip()` <br> **Output:** `hi`                                  |
| `split(sep, maxsplit)`  | Split into list                        | To split a string into a list using a separator                       | **Input:** `"a,b,c".split(",")` <br> **Output:** `['a', 'b', 'c']`                   |
| `splitlines(keepends)`  | Split by lines                         | To split a multiline string into a list of lines                      | **Input:** `"a\nb".splitlines()` <br> **Output:** `['a', 'b']`                       |
| `startswith(prefix)`    | Checks start                           | To check if a string starts with a specific prefix                    | **Input:** `"hello".startswith("he")` <br> **Output:** `True`                        |
| `zfill(width)` | Pads string on the left with zeros | To ensure numeric strings have a fixed width (useful for IDs, timestamps, formatting) | **Input:** `"42".zfill(5)` <br> **Output:** `00042` |
